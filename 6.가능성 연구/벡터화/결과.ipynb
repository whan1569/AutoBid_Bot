```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 데이터셋 생성
data = {
    '아이템명': ['사과', '배', '포도', '귤', '스마트폰', '아이스크림', '의자', '텀블러', '책', '게임기', '스피커', '스마트워치', '텔레비전', '카메라'],
    '카테고리': ['과일', '과일', '과일', '과일', '전자제품', '식품', '가구', '전자제품', '도서', '전자제품', '전자제품', '전자제품', '전자제품', '전자제품'],
    '용도/목적': ['간식, 디저트', '간식, 디저트', '간식, 디저트, 음료 재료', '간식, 비타민 C 보충', '통화, 인터넷, 앱 실행', '간식, 디저트', '앉기, 편안함', '음료수 보관, 이동용', '학습, 정보 습득', '게임 실행', '음악 재생', '시간 확인, 운동 추적', '영상 시청', '사진 촬영'],
    '물리적 형태': ['둥글고 다양한 크기, 색상', '둥글고 크기 다양함', '작은 크기, 덩어리로 모여 있음', '작은 크기, 껍질 벗기기 용이', '크기: 5.5인치, 무게: 150g', '원통형, 100g', '크기: 60x60x90cm, 무게: 3kg', '크기: 500ml, 무게: 200g', '크기: A4, 무게: 300g', '크기: 25x20x7cm, 무게: 500g', '크기: 15x15x25cm, 무게: 1kg', '크기: 4.5x4.5cm, 무게: 50g', '크기: 55인치, 무게: 8kg', '크기: 12x8x7cm, 무게: 700g'],
    '소재/구성 성분': ['천연 과일 (당분, 비타민 C, 섬유질)', '천연 과일 (수분, 당분, 비타민)', '천연 과일 (당분, 비타민 C, 섬유질)', '천연 과일 (비타민 C, 당분)', '알루미늄, 유리', '우유, 설탕, 바닐라', '나무, 패브릭', '스테인리스, 플라스틱', '종이, 잉크', '플라스틱, 전자기기', '플라스틱, 금속', '알루미늄, 고무', '유리, 플라스틱', '금속, 유리'],
}

df = pd.DataFrame(data)

# TfidfVectorizer로 텍스트 벡터화
vectorizer = TfidfVectorizer(stop_words='english')
text_data = df['카테고리'] + ' ' + df['용도/목적'] + ' ' + df['물리적 형태'] + ' ' + df['소재/구성 성분']
tfidf_matrix = vectorizer.fit_transform(text_data)

# '사과'와의 코사인 유사도 계산
target_item = '사과'
target_idx = df[df['아이템명'] == target_item].index[0]
cos_sim = cosine_similarity(tfidf_matrix[target_idx], tfidf_matrix)

# 유사도 기준으로 상위 5개 아이템 찾기
similar_items = pd.DataFrame(cos_sim.T, columns=['유사도'])
similar_items['아이템명'] = df['아이템명']
similar_items = similar_items.sort_values(by='유사도', ascending=False)

# 사과와 유사한 상위 5개 아이템 출력
similar_items.head(5)
```
